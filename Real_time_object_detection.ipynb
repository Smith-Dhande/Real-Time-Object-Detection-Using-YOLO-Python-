```python
import cv2
from ultralytics import YOLO
```

```python
obj_model = YOLO("yolov8n.pt")
vid = cv2.VideoCapture(r"C:\Users\ACER\Desktop\projects\tranining\DAY6\4K Road traffic video for object detection and tracking - free download now!.mp4")
#vid = cv2.VideoCapture(0)
```

```python
while vid.isOpened():
    ret, frame = vid.read()

    if not ret:
        print("Video not available")
        break
         # To detect the objects
    results = obj_model.predict(frame)
    # For boxes and labels over objects
    annotated_obj = results[0].plot()

    cv2.imshow("Object Detector", annotated_obj)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Exiting.....")
        break

vid.release()
cv2.destroyAllWindows()
```
